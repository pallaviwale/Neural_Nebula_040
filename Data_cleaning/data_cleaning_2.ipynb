```python
data=pd.read_csv("barcelona_weekends.csv")
data
```

```python
print(data.isnull().sum())
```

```python
data = data.drop_duplicates()
for column in data.columns:
    if data[column].dtype == 'object':  # If it's a categorical column
        data[column] = data[column].fillna('Unknown')
    else:  # If it's a numeric column
        data[column] = data[column].fillna(data[column].mean())

# Add 'city' and 'date_type' columns with specified values
data['city'] = 'Barcelona'
data['date_type'] = 'weekend'

# Save the modified dataset (optional)
data.to_csv('cleaned_and_modified_dataset.csv', index=False)  # Saves the file with the new columns

# Display the first few rows to verify
data.head() 
```
