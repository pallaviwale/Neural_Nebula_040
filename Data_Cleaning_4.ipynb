```python
import pandas as pd

# List of file names to load
file_names = [
    "paris_weekends.csv",
    "vienna_weekdays.csv",
    "vienna_weekends.csv",
    "rome_weekdays.csv",
    "rome_weekends.csv"
]

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each file name and load it into a DataFrame
for file_name in file_names:
    try:
        df = pd.read_csv(file_name)
        dataframes.append(df)  # Add the DataFrame to the list
        print(f"Successfully loaded {file_name}")
    except Exception as e:
        print(f"Error loading {file_name}: {e}")
```

```python
# Display the first few rows of each DataFrame
for i, df in enumerate(dataframes):
    print(f"\n--- Data Preview for {file_names[i]} ---")
    display(df.head())  # Show the top rows of each DataFrame
```

```python
# Clean each DataFrame
for i, df in enumerate(dataframes):
    df.dropna(inplace=True)           # Remove rows with missing values
    df.drop_duplicates(inplace=True)   # Remove duplicate rows
    df.reset_index(drop=True, inplace=True)  # Reset the index
    print(f"Cleaned data for {file_names[i]}")
```

```python
# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
print("Combined all DataFrames into a single DataFrame.")
```

```python
# Save the combined DataFrame to a new CSV file
combined_df.to_csv("combined_cleaned_data.csv", index=False)
print("Saved combined_cleaned_data.csv locally.")
```

# List of file names to load and update

```python
import pandas as pd

# List of file names to load and update
file_names = [
    "paris_weekends.csv",
    "vienna_weekdays.csv",
    "vienna_weekends.csv",
    "rome_weekdays.csv",
    "rome_weekends.csv"
]

# Loop through each file name
for file_name in file_names:
    try:
        # Load the CSV file
        df = pd.read_csv(file_name)
        
        # Extract city and day type from the file name
        city = file_name.split('_')[0].capitalize()  # Get city name and capitalize it
        day_type = file_name.split('_')[1].split('.')[0]  # Get day type (weekday or weekend)
        
        # Add new columns for city and day type
        df['city'] = city
        df['day_type'] = day_type
        
        # Save the updated DataFrame back to the same file
        df.to_csv(file_name, index=False)
        print(f"Updated and saved {file_name} with new columns.")
        
    except Exception as e:
        print(f"Error updating {file_name}: {e}")
```
